```python
import tensorflow as tf
from tensorflow import keras
import torch
import cv2
import numpy as np
from PIL import Image
from torchvision import transforms
from ultralytics import YOLO

# Constants
IMAGE_HEIGHT, IMAGE_WIDTH = 256, 256
CLASSES_LIST = ["Normal", "Shoplifting"]
HEIGHT_THRESHOLD = 250
SEQUENCE_LENGTH = 8
SLOWDOWN_FACTOR = 3  # How many times to duplicate shoplifting frames

def preprocess_frame(frame):
    resized_frame = cv2.resize(frame, (IMAGE_HEIGHT, IMAGE_WIDTH))
    return resized_frame / 255.0

def load_trained_model(model_path):
    try:
        model = keras.layers.TFSMLayer(model_path, call_endpoint="serving_default")
        print("MoviNet model loaded successfully!")
        return model
    except Exception as e:
        print(f"Error loading MoviNet model: {e}")
        return None

# def should_rotate_frame(y_coords, height):
#     """
#     Returns True if most person boxes are near the bottom (i.e., upside down).
#     """
#     if not y_coords:
#         return False
#     avg_y = np.mean(y_coords)
#     return avg_y > height * 0.75  # person is too low in the frame (likely upside-down)

def process_video(video_path, output_video_path, movienet_model):
    video_capture = cv2.VideoCapture(video_path)
    frame_rate = int(video_capture.get(cv2.CAP_PROP_FPS))
    width = int(video_capture.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(video_capture.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_video_path, fourcc, frame_rate, (width, height))

    yolov8_model = YOLO("yolov8n.pt")

    frame_buffer = []
    frame_count = 0
    rotation_decided = False
    rotate_video = False

    while video_capture.isOpened():
        ret, frame = video_capture.read()
        if not ret:
            break

        frame_count += 1
        print(f"Processing frame {frame_count}")

        # Detect persons first to check orientation
        results = yolov8_model(frame)
        persons = []
        y_positions = []

        for box, cls in zip(results[0].boxes.xyxy, results[0].boxes.cls):
            if int(cls) == 0:
                x1, y1, x2, y2 = map(int, box.cpu().numpy())
                height_box = y2 - y1
                if height_box > HEIGHT_THRESHOLD:
                    persons.append((height_box, x1, y1, x2, y2))
                    y_positions.append(y1)

        # # Decide rotation based on first few frames
        # if not rotation_decided and frame_count <= 15 and persons:
        #     rotate_video = should_rotate_frame(y_positions, height)
        #     if frame_count == 15:
        #         print(f"Auto-rotation decision: {'Rotating video 180°' if rotate_video else 'No rotation needed'}")
        #         rotation_decided = True

        # # Rotate if needed
        # if rotate_video:
        #     frame = cv2.rotate(frame, cv2.ROTATE_180)

        # Preprocess for MoviNet
        processed_frame = preprocess_frame(frame)
        frame_buffer.append(processed_frame)
        if len(frame_buffer) > SEQUENCE_LENGTH:
            frame_buffer.pop(0)

        predicted_label = "Normal"
        predicted_prob = 0.0

        if len(frame_buffer) == SEQUENCE_LENGTH:
            input_data = np.expand_dims(np.array(frame_buffer), axis=0)
            predictions = movienet_model(input_data)
            predictions = list(predictions.values())[0].numpy()[0]
            predicted_class = np.argmax(predictions)
            predicted_label = CLASSES_LIST[predicted_class]
            predicted_prob = round(float(predictions[predicted_class]), 2)

            print(f"[MoviNet] Predicted: {predicted_label} ({predicted_prob:.2f})")

        # Set color
        color = (0, 255, 0) if predicted_label == "Normal" else (0, 0, 255)

        for _, x1, y1, x2, y2 in persons:
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, predicted_label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

        overlay_text = f"Prediction: {predicted_label} ({predicted_prob:.2f})"
        cv2.putText(frame, overlay_text, (10, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

        # ✅ Slow down on shoplifting
        if predicted_label == "Shoplifting":
            for _ in range(SLOWDOWN_FACTOR):
                out.write(frame)
        else:
            out.write(frame)


    frame = cv2.rotate(frame, cv2.ROTATE_180)
    video_capture.release()
    out.release()
    cv2.destroyAllWindows()
    print(f"✅ Processing complete. Output saved to: {output_video_path}")

# Example usage
movienet_model = load_trained_model("/content/drive/MyDrive/MOVINET_SHOPLIFTING_AllahBad")
process_video(
    "/content/drive/MyDrive/FYP Dataset/Shoplifting/Shoplifting-2.mp4",
    "/content/drive/MyDrive/bounding_box_folder_movienet_Allahbad/Shoplifting-2_output_video_cmpds.mp4",
    movienet_model
)
```
